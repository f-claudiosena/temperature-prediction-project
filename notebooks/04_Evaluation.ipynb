```python
# 04_Evaluation.ipynb
import joblib
import matplotlib.pyplot as plt

model = joblib.load('/workspaces/temperature-prediction-project/data/random_forest_model.pkl')
_, X_test, _, y_test, _ = joblib.load('/workspaces/temperature-prediction-project/data/preprocessed_data.pkl')

y_pred = model.predict(X_test)

# Grfico de comparao
plt.figure(figsize=(10, 5))
plt.scatter(y_test, y_pred, alpha=0.3)
plt.xlabel("Valor Real")
plt.ylabel("Previso")
plt.title("Real vs. Previso de Temperatura")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()
```
