```python
# 01_EDA.ipynb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataframe load
df = pd.read_csv('/workspaces/temperature-prediction-project/data/weatherHistory.csv')

# Infos
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())

# Descriptive statistics
print(df.describe())

# Date conversion
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)

# Extra feature: hour and month
df['hour'] = df['Formatted Date'].dt.hour
df['month'] = df['Formatted Date'].dt.month
```

# Graphics

```python
sns.histplot(df['Temperature (C)'], kde=True, bins=50)
plt.title("Temperature Distribution")
plt.xlabel("Temperature (°C)")  
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Precip Type', y='Temperature (C)', data=df)
plt.title("Temperature by Precip Type")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlations between Numerical Variables")
plt.show()
```
